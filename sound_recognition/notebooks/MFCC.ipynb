```python
fs = 8820
time = np.linspace(0, 1.0, fs)
sinwav1 = 1.2 * np.sin(2 * np.pi * 130 * time)
coswav1 = 1.8 * np.cos(2 * np.pi * 200 * time)
sinwav2 = 0.8 * np.sin(2 * np.pi * 260 * time)
coswav2 = 2.3 * np.cos(2 * np.pi * 320 * time)
wav = 1.4 + (sinwav1 + coswav1 + sinwav2 + coswav2)

plt.plot(time * 1000, wav)
plt.xlabel("time [ms]")
plt.ylabel("amplitude")
plt.xlim(0, 50)
plt.show()
```
