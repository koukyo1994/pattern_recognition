```julia
using(Distributions)
using(GLM)
```

```julia
dist = Normal(0.0, 0.05)
eps = rand(dist, 100)
x = collect(0.01:0.01:1)
f(x) = 0.5 + 0.4 .* sin(2pi.*x) .+ eps
```

```julia
using Plots
```

```julia
gr()
plot(x, f(x))
```

```julia
scatter(x, f(x))
```

```julia
index = collect(1:10:91)
x_i = x[index]
f_i = f(x)[index]
```

```julia
using DataFrames
```

```julia
sample = DataFrame(x = x_i, f_x = f_i)
```

```julia
lm = fit(LinearModel, f_i~x_i, sample)
```
